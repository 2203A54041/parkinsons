```python
def outlier_thresholds(dataframe, col_name, q1=0.10, q3=0.90):
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

def check_outlier(dataframe, col_name):
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False

def replace_with_thresholds(dataframe, variable, q1=0.05, q3=0.95):
    low_limit, up_limit = outlier_thresholds(dataframe, variable, q1=0.05, q3=0.95)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit
```

```python
def hloa(objective_func, bounds, pop_size, iterations):
    """
    Horned Lizard Optimization Algorithm (HLOA)

    Args:
        objective_func: The function to be minimized.
        bounds: A list of tuples, where each tuple represents the lower and upper bound for a dimension.
        pop_size: The number of horned lizards in the population.
        iterations: The number of iterations to run the algorithm.

    Returns:
        The best solution found and its corresponding objective function value.
    """
    dim = len(bounds)
    # Initialize population
    population = np.random.uniform([b[0] for b in bounds], [b[1] for b in bounds], size=(pop_size, dim))
    fitness = np.array([objective_func(p) for p in population])

    best_solution = population[np.argmin(fitness)]
    best_fitness = np.min(fitness)

    for iter in range(iterations):
        new_population = np.copy(population)
        for i in range(pop_size):
            # Foraging behavior
            r1 = np.random.rand()
            if r1 < 0.5:
                new_population[i] = population[i] + np.random.rand() * (best_solution - population[i])
            # Defensive behavior
            else:
                rand_lizard = population[np.random.randint(pop_size)]
                new_population[i] = population[i] + np.random.rand() * (rand_lizard - population[i])

            # Ensure bounds are respected
            for d in range(dim):
                new_population[i][d] = np.clip(new_population[i][d], bounds[d][0], bounds[d][1])

        # Reproduction behavior
        sorted_indices = np.argsort(fitness)
        best_lizard = population[sorted_indices[0]]
        worst_lizard = population[sorted_indices[-1]]

        for i in range(pop_size):
            r2 = np.random.rand()
            if r2 < 0.5:
                new_solution = best_lizard + np.random.rand() * (best_lizard - population[i])
            else:
                new_solution = worst_lizard + np.random.rand() * (best_lizard - worst_lizard)

            # Ensure bounds are respected
            for d in range(dim):
                new_solution[d] = np.clip(new_solution[d], bounds[d][0], bounds[d][1])

            new_population = np.vstack([new_population, new_solution])


        # Evaluate fitness and update population
        new_fitness = np.array([objective_func(p) for p in new_population])
        combined_population = np.vstack([population, new_population])
        combined_fitness = np.concatenate([fitness, new_fitness])

        sorted_indices = np.argsort(combined_fitness)
        population = combined_population[sorted_indices[:pop_size]]
        fitness = combined_fitness[sorted_indices[:pop_size]]

        # Update best solution
        if fitness[0] < best_fitness:
            best_fitness = fitness[0]
            best_solution = population[0]

    return best_solution, best_fitness
```

```python
def objective_func(params):
    """
    Objective function for HLOA to minimize.

    Args:
        params: A list or array of hyperparameters for the RandomForestClassifier.

    Returns:
        1 - accuracy of the RandomForestClassifier model.
    """
    # Hyperparameters for RandomForestClassifier
    n_estimators = int(params[0])
    max_depth = int(params[1]) if params[1] > 0 else None
    min_samples_split = max(2, int(params[2]))
    min_samples_leaf = max(1, int(params[3]))

    # Instantiate the model with the hyperparameters
    model = RandomForestClassifier(n_estimators=n_estimators,
                                   max_depth=max_depth,
                                   min_samples_split=min_samples_split,
                                   min_samples_leaf=min_samples_leaf,
                                   random_state=42)

    # Train the model
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)

    # Return 1 - accuracy for minimization
    return 1 - accuracy
```

```python
model_bounds = {
    'LR': [(-10.0, 10.0)], # Example bound for C in Logistic Regression (log scale)
    'KNN': [(1, 30)], # Example bound for n_neighbors in KNN
    'CART': [(1, 20), (2, 20)], # Example bounds for max_depth and min_samples_split in Decision Tree
    'RF': [(50, 300), (1, 30), (2, 20), (1, 20)], # Example bounds for n_estimators, max_depth, min_samples_split, min_samples_leaf in Random Forest
    'SVC': [(-10.0, 10.0), (-10.0, 10.0)], # Example bounds for C and gamma in SVC (log scale)
    'GBM': [(50, 300), (0.01, 0.2)], # Example bounds for n_estimators and learning_rate in Gradient Boosting
    "XGBoost": [(50, 300), (0.01, 0.2)], # Example bounds for n_estimators and learning_rate in XGBoost
    "LightGBM": [(50, 300), (0.01, 0.2)], # Example bounds for n_estimators and learning_rate in LightGBM
    "CatBoost": [(50, 300), (0.01, 0.2)] # Example bounds for n_estimators and learning_rate in CatBoost
}
```

```python
optimized_models = {}

for name, model in models:
    print(f"Optimizing {name}...")
    if name in model_bounds:
        bounds = model_bounds[name]

        def model_objective_func(params):
            # Instantiate the model with the hyperparameters based on model type
            if name == 'LR':
                model_instance = LogisticRegression(C=10**params[0], random_state=42)
            elif name == 'KNN':
                model_instance = KNeighborsClassifier(n_neighbors=int(params[0]))
            elif name == 'CART':
                max_depth = int(params[0]) if params[0] > 0 else None
                min_samples_split = max(2, int(params[1]))
                model_instance = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, random_state=42)
            elif name == 'RF':
                n_estimators = int(params[0])
                max_depth = int(params[1]) if params[1] > 0 else None
                min_samples_split = max(2, int(params[2]))
                min_samples_leaf = max(1, int(params[3]))
                model_instance = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf, random_state=42)
            elif name == 'SVC':
                 model_instance = SVC(C=10**params[0], gamma=10**params[1], random_state=42)
            elif name == 'GBM':
                 model_instance = GradientBoostingClassifier(n_estimators=int(params[0]), learning_rate=params[1], random_state=42)
            elif name == 'XGBoost':
                 model_instance = XGBClassifier(n_estimators=int(params[0]), learning_rate=params[1], random_state=42, use_label_encoder=False, eval_metric='logloss')
            elif name == 'LightGBM':
                 model_instance = LGBMClassifier(n_estimators=int(params[0]), learning_rate=params[1], random_state=42, verbose=-1)
            elif name == 'CatBoost':
                 model_instance = CatBoostClassifier(n_estimators=int(params[0]), learning_rate=params[1], random_state=42, verbose=False)
            else:
                raise ValueError(f"Unknown model: {name}")


            model_instance.fit(X_train, y_train)
            y_pred = model_instance.predict(X_test)
            accuracy = accuracy_score(y_test, y_pred)
            return 1 - accuracy

        best_params, best_fitness = hloa(model_objective_func, bounds, pop_size=20, iterations=10) # Adjusted pop_size and iterations for quicker execution

        print(f"Best parameters for {name}: {best_params}")
        print(f"Best fitness (1 - accuracy) for {name}: {best_fitness}")

        # Instantiate the model with the best hyperparameters
        if name == 'LR':
            optimized_model = LogisticRegression(C=10**best_params[0], random_state=42)
        elif name == 'KNN':
            optimized_model = KNeighborsClassifier(n_neighbors=int(best_params[0]))
        elif name == 'CART':
            max_depth = int(best_params[0]) if best_params[0] > 0 else None
            min_samples_split = max(2, int(best_params[1]))
            optimized_model = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split, random_state=42)
        elif name == 'RF':
            n_estimators = int(best_params[0])
            max_depth = int(best_params[1]) if best_params[1] > 0 else None
            min_samples_split = max(2, int(best_params[2]))
            min_samples_leaf = max(1, int(best_params[3]))
            optimized_model = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf, random_state=42)
        elif name == 'SVC':
             optimized_model = SVC(C=10**best_params[0], gamma=10**best_params[1], random_state=42)
        elif name == 'GBM':
             optimized_model = GradientBoostingClassifier(n_estimators=int(best_params[0]), learning_rate=best_params[1], random_state=42)
        elif name == 'XGBoost':
             optimized_model = XGBClassifier(n_estimators=int(best_params[0]), learning_rate=best_params[1], random_state=42, use_label_encoder=False, eval_metric='logloss')
        elif name == 'LightGBM':
             optimized_model = LGBMClassifier(n_estimators=int(best_params[0]), learning_rate=best_params[1], random_state=42, verbose=-1)
        elif name == 'CatBoost':
             optimized_model = CatBoostClassifier(n_estimators=int(best_params[0]), learning_rate=best_params[1], random_state=42, verbose=False)
        else:
            raise ValueError(f"Unknown model: {name}")

        optimized_model.fit(X_train, y_train)
        optimized_models[name] = optimized_model

    else:
        print(f"No bounds defined for {name}. Training with default hyperparameters.")
        model.fit(X_train, y_train)
        optimized_models[name] = model
```

*   Hyperparameter bounds were defined for Logistic Regression, K-Nearest Neighbors, Classification and Regression Trees, Random Forest, Support Vector Classifier, Gradient Boosting Machines, XGBoost, LightGBM, and CatBoost.
